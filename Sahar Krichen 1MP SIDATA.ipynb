```python
age_bins = [18, 30, 40, 50, 60, 70, 80, 90]
age_labels = ['18-29', '30-39', '40-49', '50-59', '60-69', '70-79', '80-90']
df['AgeGroup'] = pd.cut(df['age'], bins=age_bins, labels=age_labels)
age_income_counts = df.groupby(['AgeGroup', 'income <=50k'])['income <=50k'].count().unstack()
fig, ax = plt.subplots(len(age_labels), figsize=(10, 20))
for i, age_label in enumerate(age_labels):
    age_income_counts.loc[age_label].plot(kind='bar', ax=ax[i], color=['blue', 'red'])
    ax[i].set_title(f'Age Group: {age_label}')
    ax[i].set_xlabel('income <=50k')
    ax[i].set_ylabel('Count')
    ax[i].legend(['>50K', '<=50K'])

plt.tight_layout()
plt.show()
```

```python
race_income = df.groupby(['race', 'income <=50k'])['income <=50k'].count().unstack()
race_income['Total'] = race_income.sum(axis=1)
race_income['% >50K'] = round(race_income[0] / race_income['Total'] * 100, 2)
race_income = race_income.reset_index()
fig, ax = plt.subplots(figsize=(10,6))
x = race_income['race']
y1 = race_income['% >50K']
y2 = 100 - y1
ax.bar(x, y1, label='>50K', color='g')
ax.bar(x, y2, bottom=y1, label='<=50K', color='r')
ax.set_xlabel('Race')
ax.set_ylabel('Pourcentage')
ax.set_title('Pourcentage de revenu >50K par rapport la Race')
ax.legend()
plt.show()
```
